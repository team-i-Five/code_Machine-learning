```python
import os # 운영 체제 관련 작업을 수행하기 위한 Python 표준 라이브러리 중 하나 -> 자바 환경 변수를 설정하기 위해 사용
import pandas as pd # 데이터 조작 및 분석을 위한 파이썬 라이브러리 -> 데이터를 데이터프레임으로 로드하고 조작하는 데 사용
import numpy as np # 배열 및 수치 연산을 수행하기 위한 라이브러리 -> 다차원 배열 및 수학적 함수를 다루는 데 사용
from konlpy.tag import Okt # Konlpy 라이브러리에서 제공하는 한국어 형태소 분석기 중 하나 ->  Okt를 초기화하는 데 사용
from gensim.models.word2vec import Word2Vec # Gensim 라이브러리에서 제공하는 Word2Vec 모델을 생성하기 위한 클래스 -> 단어를 고차원 벡터로 표현하여 단어 간 유사성 및 의미 관계를 파악
```

```python
# 데이터를 데이터프레임으로 로드
vector_data = pd.read_csv('../musical_data_vector_date.csv', encoding='utf-8')# 데이터 파일 경로
vector_data
```

### 저장한 학습모델 불러오기

```python
model = Word2Vec.load("word2vec_model2.bin")
```

```python
#Vector 형태로 변환한 단어 리스트 조회
model.wv.key_to_index
```

### 단어를 조회하여 종류 8가지 선정완료

```python
# "tokenized_data" 열에서 품사별로 분리된 단어 리스트를 추출
tokenized_data = vector_data['tokenized_data'].apply(eval)
tokenized_data
```

```python
# 장르와 유사한 단어를 찾음
selected_genre = ["웃음", "슬픔", "사랑", "마법", "공포", "역사", "오페라", "예술"]
similar_words_by_genre = {}

for genre in selected_genre:
    similar_words = model.wv.most_similar(genre, topn=100)
    similar_words_by_genre[genre] = similar_words
similar_words_by_genre
```

```python
# '장르' 열 초기화
vector_data['장르'] = None
```

```python
# 각 작품에 대해 유사한 단어를 확인하고, 해당 장르 열에 값을 할당
for index, row in vector_data.iterrows():
    synopsis = row['synopsis']
    assigned_genre = None
    
    for genre in selected_genre:
        if not pd.isna(synopsis) and genre in synopsis:
            assigned_genre = genre
            break  # 장르를 찾았으면 종료

    if assigned_genre is None and not pd.isna(synopsis):
        # 시놉시스가 존재하고 장르가 아직 할당되지 않았다면
        # 유사한 단어를 사용하여 장르를 찾기
        max_similarity = 0
        for genre, similar_words in similar_words_by_genre.items():
            for word, similarity in similar_words:
                if word in synopsis and similarity > max_similarity:
                    assigned_genre = genre
                    max_similarity = similarity

    if assigned_genre:
        vector_data.at[index, '장르'] = assigned_genre

# 결과 확인
print(vector_data[['title', 'synopsis', '장르']])
```
