```python
# 데이터를 데이터프레임으로 로드
vector_data = pd.read_csv('../book_data_vector.csv', encoding='utf-8')# 데이터 파일 경로
vector_data
```

### 저장한 학습모델 불러오기

```python
model = Word2Vec.load("word2vec_model.bin")
```

```python
#Vector 형태로 변환한 단어 리스트 조회
model.wv.key_to_index
```

### 단어를 조회하여 종류 10가지 선정완료

```python
# "tokenized_data" 열에서 품사별로 분리된 단어 리스트를 추출
tokenized_data = vector_data['tokenized_data'].apply(eval)
tokenized_data
```

```python
# 선택한 장르 (예: "사랑")와 유사한 단어를 찾음
selected_genre = "사랑"
similar_words = model.wv.most_similar(selected_genre, topn=100)
similar_words
```

### 유사도가 높은 100개의 단어를 하나의 종류로 통합하기

```python
from collections import Counter
```

```python
# "tokenized_data" 열에서 토큰 리스트를 모두 합치고 공백으로 분리하여 단어 리스트 생성
words = " ".join(tokenized_data.apply(lambda tokens: " ".join(tokens))).split()
```

```python
# 단어 빈도 집계
word_counts = Counter(words)
```

```python
# 빈도가 가장 높은 단어 상위 200개를 출력
most_common_words = word_counts.most_common(200)
for idx, (word, count) in enumerate(most_common_words, 1):
    print(f"순위: {idx}, 단어: {word}, 빈도: {count}")
```
