```python
import os # 운영 체제 관련 작업을 수행하기 위한 Python 표준 라이브러리 중 하나 -> 자바 환경 변수를 설정하기 위해 사용
import pandas as pd # 데이터 조작 및 분석을 위한 파이썬 라이브러리 -> 데이터를 데이터프레임으로 로드하고 조작하는 데 사용
import numpy as np # 배열 및 수치 연산을 수행하기 위한 라이브러리 -> 다차원 배열 및 수학적 함수를 다루는 데 사용
from konlpy.tag import Okt # Konlpy 라이브러리에서 제공하는 한국어 형태소 분석기 중 하나 ->  Okt를 초기화하는 데 사용
from gensim.models.word2vec import Word2Vec # Gensim 라이브러리에서 제공하는 Word2Vec 모델을 생성하기 위한 클래스 -> 단어를 고차원 벡터로 표현하여 단어 간 유사성 및 의미 관계를 파악
```

```python
# 데이터를 데이터프레임으로 로드
vector_data = pd.read_csv('../mock_test_csv/musical_data_vector_date.csv', encoding='utf-8')# 데이터 파일 경로
vector_data
```

### 저장한 학습모델 불러오기

```python
model = Word2Vec.load("word2vec_model2.bin")
```

```python
#Vector 형태로 변환한 단어 리스트 조회
model.wv.key_to_index
```

### 단어를 조회하여 종류 10가지 선정완료

```python
# "tokenized_data" 열에서 품사별로 분리된 단어 리스트를 추출
tokenized_data = vector_data['tokenized_data'].apply(eval)
tokenized_data
```

```python
# 장르와 유사한 단어를 찾음
selected_genre = ["웃음","친구","아이","사람", "슬픔", "사랑", "마법", "공포", "역사", "오페라", "예술"] # 범주화할 장르10가지 list에 저장
similar_words_by_genre = {} # 장르별로 유사한 단어를 키-값 쌍으로 저장하기 위해 딕셔너리 초기화

for genre in selected_genre:
    similar_words = model.wv.most_similar(genre, topn=100) # 모델에서 장르별 단어와 가장 유사한 단어 100개를 similar_words에 저장
    similar_words_by_genre[genre] = similar_words # 장르와 유사한단어를 해당 장르의 키로 저장
similar_words_by_genre
```

### 유사도가 높은 100개의 단어를 하나의 종류로 통합하기
[정보] 빈도수 0이 가장 빈도수 높은 단어

코미디 -> '웃음': 890

'슬픔': 315

로맨스 -> '사랑': 4

판타지 -> '마법': 469

스릴러 -> '공포': 829

역사 -> '과거': 249

'오페라': 964

예술 -> '예술': 550


```python
# '장르' 열 초기화
vector_data['genre'] = None
vector_data
```

```python
# 각 작품에 대해 유사한 단어를 확인하고, 해당 장르 열에 값을 할당
# iterrows() : pandas에서 DataFrame을 반복하는데 사용되는 메서드
# isna : 주어진 값이 NaN인지 여부 확인을 위해 사용되는 pandas라이브러리에서 사용되는 함수
for index, row in vector_data.iterrows(): # 데이터프레임의 값을 index값과 row별로 나눠서 반복
    synopsis = row['synopsis'] # 각 행의 줄거리를 synopsis에 저장
    assigned_genre = None # 각행에 해당하는 장르를 저장할 변수 초기화
    
    for genre in selected_genre:
        if not pd.isna(synopsis) and genre in synopsis: # 줄거리가 NaN값이 아니고 장르 모두에 줄거기가 있다면
            assigned_genre = genre # 장르를 assigned_genre에 저장
            break  # 장르저장 후 종료

    if assigned_genre is None and not pd.isna(synopsis):
        # 만약 장르가 아직 할당되지 않았고 시놉시스(줄거리)가 존재한다면 아래 작업 수행
        # 유사한 단어를 사용하여 장르를 찾기
        max_similarity = 0 # 유사도의 최댓값을 저장하는 변수 초기화
        for genre, similar_words in similar_words_by_genre.items(): # 'similar_words_by_genre' 딕셔너리에서 각 장르와 해당 장르의 유사한 단어 리스트를 가져옴 - group by생각하기
            for word, similarity in similar_words: # 각 장르의 유사한 단어 리스트에서 단어와 유사도를 가져옴
                if word in synopsis and similarity > max_similarity: # 만약 해당 단어가 시놉시스(줄거리)에 포함되고 유사도가 최댓값보다 크다면
                    assigned_genre = genre # 장르로 할당
                    max_similarity = similarity # 최댓값을 현재 유사도로 업데이트

    if assigned_genre:
        vector_data.at[index, 'genre'] = assigned_genre # 데이터프레임에 저장 -> 해당인엑스의 장르열에 장르저장

# 결과 확인
print(vector_data[['title', 'synopsis', 'genre']])
```
