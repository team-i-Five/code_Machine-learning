```python
import os # 운영 체제 관련 작업을 수행하기 위한 Python 표준 라이브러리 중 하나 -> 자바 환경 변수를 설정하기 위해 사용
import pandas as pd # 데이터 조작 및 분석을 위한 파이썬 라이브러리 -> 데이터를 데이터프레임으로 로드하고 조작하는 데 사용
import numpy as np # 배열 및 수치 연산을 수행하기 위한 라이브러리 -> 다차원 배열 및 수학적 함수를 다루는 데 사용
from konlpy.tag import Okt # Konlpy 라이브러리에서 제공하는 한국어 형태소 분석기 중 하나 ->  Okt를 초기화하는 데 사용
from gensim.models.word2vec import Word2Vec # Gensim 라이브러리에서 제공하는 Word2Vec 모델을 생성하기 위한 클래스 -> 단어를 고차원 벡터로 표현하여 단어 간 유사성 및 의미 관계를 파악
```

```python
# 데이터를 데이터프레임으로 로드
vector_data = pd.read_csv('../mock_test_csv/musical_data_vector2.csv', encoding='utf-8')# 데이터 파일 경로
vector_data
```

### 저장한 학습모델 불러오기

```python
model = Word2Vec.load("word2vec_model2.bin")
```

```python
#Vector 형태로 변환한 단어 리스트 조회
model.wv.key_to_index
```

### 단어를 조회하여 종류 10가지 선정완료

```python
# "tokenized_data" 열에서 품사별로 분리된 단어 리스트를 추출
tokenized_data = vector_data['tokenized_data'].apply(eval)
tokenized_data
```

```python
# 장르와 유사한 단어를 찾음
selected_genre = ["웃음","친구","아이","사람", "슬픔", "사랑", "마법", "공포", "역사", "오페라", "예술"] # 범주화할 장르10가지 list에 저장
similar_words_by_genre = {} # 장르별로 유사한 단어를 키-값 쌍으로 저장하기 위해 딕셔너리 초기화

for genre in selected_genre:
    similar_words = model.wv.most_similar(genre, topn=100) # 모델에서 장르별 단어와 가장 유사한 단어 100개를 similar_words에 저장
    similar_words_by_genre[genre] = similar_words # 장르와 유사한단어를 해당 장르의 키로 저장
```

### 유사도가 높은 100개의 단어를 하나의 종류로 통합하기
[정보] 빈도수 0이 가장 빈도수 높은 단어

코미디 -> '웃음': 890

'슬픔': 315

로맨스 -> '사랑': 4

판타지 -> '마법': 469

스릴러 -> '공포': 829

역사 -> '과거': 249

'오페라': 964

예술 -> '예술': 550


```python
# 'tag' 열 초기화
for i in range(1, 4):
    vector_data[f'tag{i}'] = None
vector_data
```

```python
# 장르 할당
for index, row in vector_data.iterrows():
    synopsis = row['synopsis']
    assigned_genres = set()  # 중복을 허용하지 않는 세트로 변경

    for genre in selected_genre:
        if not pd.isna(synopsis) and genre in synopsis:
            assigned_genres.add(genre)  # 세트에 장르 추가

    if len(assigned_genres) < 3 and not pd.isna(synopsis):
        max_similar_genres = set()
        max_similarity = 0

        for genre, similar_words in similar_words_by_genre.items():
            if genre not in assigned_genres:  # 이미 태그로 지정된 장르가 아닌 경우에만 처리
                for word, similarity in similar_words:
                    if word in synopsis and similarity > max_similarity:
                        max_similar_genres = {genre}
                        max_similarity = similarity

        assigned_genres.update(max_similar_genres)

    assigned_genres = list(assigned_genres)[:3]  # 최대 3개까지만 저장

    for i, genre in enumerate(assigned_genres):
        vector_data.at[index, f'tag{i + 1}'] = genre

    # tag2와 tag3에 NaN 값 설정
    if len(assigned_genres) < 2:
        vector_data.at[index, 'tag2'] = 'NaN'
    if len(assigned_genres) < 3:
        vector_data.at[index, 'tag3'] = 'NaN'

# 결과 확인
print(vector_data[['title', 'synopsis', 'tag1', 'tag2', 'tag3']])
```
